```python
import pandas as pd 
import numpy as np 
```

```python
quantidade_produtos = [3,5,6,10]
precos_produtos = [10,5,7,100]

array_produtos = np.array(quantidade_produtos)
array_precos = np.array(precos_produtos)

faturamento = array_produtos * array_precos
print(faturamento)
```

```python
quantidade_produtos = [3,5,6,10]
precos_produtos = [10,5,7,100]
nomes_produtos =["Pastel", "Caçulinha", "Refrigerante", "Combos"]

array_produtos = np.array(quantidade_produtos)
array_precos = np.array(precos_produtos)

faturamento = array_produtos * array_precos
for nome,preco in zip(nomes_produtos,faturamento):
    print(f"{nome} : {preco:.2f}")
```

### Exercício de fixação 
- Criar uma lista de irá possuir 10 nomes de produtos.
- Criar uma lista que irá possuir o preço de custo(atacado).
- Criar uma lista que irá possuir o preço de venda(varejo).
- Ao final, exibir o lucro obtido em cada produto.


```python
nomes_produtos =["Samsung A71", "Samsung S21", "Samsung J7 prime", "Iphone XR", "Xiaomi mi 11", "Redmi note 10 pro", "Samsung Z fold 4", "Iphone 11", "Samsung A33", "Samsung tab A7"]
precos_produtos = [600, 1500, 400, 750, 1100, 700, 3800, 1100, 700, 600]
custo_produtos =  [200, 500, 100, 350, 300, 250, 1000, 500, 250, 200]

array_precos = np.array(precos_produtos)
array_custo = np.array(custo_produtos)
faturamento = array_precos - array_custo
for nome,preco in zip(nomes_produtos,faturamento):
    print(f"Lucro de produtos vendidos: {nome} = {preco:.2f}")
faturamento_total = np.sum(faturamento)
print(f"Seu lucro total foi de {faturamento_total:.2f} ")
  
```
