```python
#Importação de bibliotecas
import numpy as np #np será o apelido 
```

```python
temperaturas = np.array([30,40,50,60,80,90,100])
print(temperaturas)
print(type(temperaturas))
```

```python
#Media aritmética
media = np.mean(temperaturas)
print(f"{media:.2f}")
```

```python
#retorna o maior valor da lista
maior_valor = np.max(temperaturas)
print(maior_valor)
```

```python
#retorna o menor valor da lista
menor_valor = np.min(temperaturas)
print(menor_valor)
```

```python
#Salários x 1000
salarios = np.array([2,3,4,5,1,500])
media_salarios = np.mean(salarios)
print(f"{media_salarios:.2f}")
```

```python
mediana = np.median(salarios)
print(mediana)
```

Desafio Final:
- Criar um array de habitantes de 5 cidades do Estado de SP.
- Cidade que possui maior população 
- Cidade que possui menor população
- Média da população 
- Mediana


```python
habitantes = np.array([269.657, 15.718, 247.500, 158.647, 145.155])
print(f"Indaiatuba habitantes = {habitantes[0]}, Holambra = {habitantes[1]}, Americana = {habitantes[2]}, Atibaia= {habitantes[3]} Bocutacu = {habitantes[4]}")
menor_popul = np.min(habitantes)
print(f"A menor população é {menor_popul}")
maior_popul = np.max(habitantes)
print(f"A maior população é {maior_popul}")
media_popul = np.mean(habitantes)
print(f"A média da população é {media_popul}")
mediana_popul = np.median(habitantes)
print(f"A mediana da população é {mediana_popul}")
```
