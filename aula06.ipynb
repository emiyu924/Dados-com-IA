### Ao analisar as informações fornecidas pelo código acima, notamos que a categoria: "parental_education_level" tem cerca de 91 espaços na tabela "nulos" ou "vazios". Usamos os comandos **not null** e **isna** para detectarmos quantos são nulos e quantos não são.

## Criação de coluna com os dados da tabela 

```python
colunas = [
    "age",
    "study_hours_per_day",
    "social_media_hours", 
    "netflix_hours",
    "attendance_percentage",
    "sleep_hours",
    "exercise_frequency",
    "mental_health_rating",
    "exam_score"
]
```

### Gráfico criado para análise geral da relação de cada coluna entre si
*juntamente com a escolha de cores do conjunto: "coolwarm" para destaque e analogia com cores azul e vermelho para "quente" e "frio"* 

```python
sns.heatmap(data=df[colunas].corr(),annot=True, cmap="coolwarm", fmt=".2f");
```

## Gráfico da relação entre a nota do exame com a quantidade de horas de estudo 
*Observe que foram feitos 2 gráficos (apenas invertendo o eixo x e y) para análise de formatos (o primeiro acaba tendo um visual um pouco mais confuso e mais complicado para visualização)*

```python
sns.lmplot(data=df, y="study_hours_per_day", x="exam_score")
```

```python
sns.lmplot(data=df, x="study_hours_per_day", y="exam_score")
```

## Gráfico da relação entre a nota do exame com o horas assistidas do netflix 
*Como pode ser observado por não ter uma relação forte o gráfico fica visualmente bem espalhado e bagunçado*

```python
sns.lmplot(data=df, x="netflix_hours", y="exam_score")
```

## Gráfico da relação entre "horas estudadas" e "avaliação da saúde mental"

```python
sns.lmplot(data=df, x="study_hours_per_day", y="mental_health_rating")
```

## Filtrando pela média dos alunos com relação as horas estudadas


```python
filtro_horas_altas = df["study_hours_per_day"] >= 5
filtro_horas_baixas = df["study_hours_per_day"] < 2

media_alta = df[filtro_horas_altas]["exam_score"].mean()
media_baixa =df[filtro_horas_baixas]["exam_score"].mean()

print(f"A média das pessoas que estudaram 5 ou mais horas foi {media_alta:.2f}, já a média das pessoas que estudaram menos de duas foras foi de {media_baixa:.2f}")
```
