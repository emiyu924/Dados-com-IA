```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(r".\data\base_estudantes_tratado.csv")
```

```python
sns.lmplot(data=df, x="study_hours_per_day", y="exam_score");
```

```python
X = df[["study_hours_per_day","social_media_hours","netflix_hours","attendance_percentage","sleep_hours","exercise_frequency","mental_health_rating"]]
y = df["exam_score"]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
modelo = LinearRegression()

modelo.fit(X,y)
```
