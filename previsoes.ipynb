```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(r".\data\base_estudantes_tratado.csv")
```

```python
sns.lmplot(data=df, x="study_hours_per_day", y="exam_score");
```

```python
X = df[["study_hours_per_day", "social_media_hours",]]
y = df["exam_score"]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
modelo = LinearRegression()

modelo.fit(X,y)
```

```python
novo_aluno = pd.DataFrame({"study_hours_per_day":[1]})
modelo.predict(novo_aluno)
```

```python
modelo.coef_
```

```python
modelo.intercept_
```

```python
modelo.score(X,y)
```
