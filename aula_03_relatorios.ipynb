**Resposta da IA:**

Análise dos hábitos que mais impactam a nota dos alunos

Com base na análise da base de dados com 1.000 alunos, os três hábitos que apresentam maior impacto na nota final (exam_score) são:

Horas de estudo por dia – Correlação: +0,825

Este é o fator que mais influencia o desempenho dos alunos. A forte correlação positiva indica que, quanto maior o tempo dedicado aos estudos diariamente, maiores tendem a ser as notas obtidas no exame.

Avaliação da saúde mental – Correlação: +0,322

A saúde mental também exerce influência significativa no desempenho acadêmico. Alunos com melhor bem-estar emocional apresentam, em média, resultados superiores nas avaliações.

Horas assistindo Netflix – Correlação: −0,172

O tempo gasto assistindo Netflix apresenta uma relação negativa com a nota. Isso significa que alunos que dedicam mais horas a essa atividade tendem a obter notas um pouco menores.

Conclusão para a direção da escola

Os resultados mostram que o principal fator associado ao bom desempenho dos alunos é o tempo de estudo diário, com uma correlação de 0,825, muito superior aos demais hábitos analisados. Além disso, promover ações voltadas ao bem-estar e à saúde mental dos estudantes pode contribuir para melhores resultados acadêmicos. Por outro lado, é recomendável orientar os alunos sobre o uso equilibrado do tempo destinado ao entretenimento, como assistir séries, para que essa atividade não prejudique sua rotina de estudos.

Apesar da diferença entre os grupos, ela é pequena (cerca de 2,3 pontos entre a maior e a menor média), indicando que a alimentação pode influenciar o desempenho, mas não é um dos fatores mais determinantes nesta base.
