# Análise de Fatores que Influenciam o Desempenho Acadêmico

> Base: **1.000 estudantes universitários** — variáveis: qualidade da dieta, trabalho de meio período, escolaridade dos pais, horas de sono e nota no exame (`exam_score`).

---

## 1. 🍎 Qualidade da dieta vs. nota

A diferença existe, mas é **modesta**. Quem se alimenta pior tira em média ~2 pontos a menos.
 



### Análise estatística: A diferença entre o grupo de dieta "Fair" e "Poor" é de ~2,3 pontos. Embora exista um gradiente (quem se alimenta pior vai pior), a magnitude é modesta.

### Como nutricionista e psicólogo: Uma dieta ruim afeta cognição (deficiência de ferro, ômega-3, vitaminas do complexo B prejudica memória e atenção), mas na prática universitária o efeito aparece diluído porque outros fatores (sono, estresse, ansiedade) pesam mais. A mensagem é: não é o fator dominante, mas importa — especialmente para adolescentes em fase de desenvolvimento cerebral.

## 2. 💼 Alunos com trabalho de meio período vão pior?






### Teste t: t = 0,85, p = 0,39 → diferença NÃO significativa.

### Interpretação: Contraintuitivamente, ter trabalho de meio período não reduz significativamente a nota nesta amostra. Isso pode refletir um efeito de seleção: quem consegue conciliar trabalho e estudo tende a ser mais organizado, disciplinado e resiliente — compensando a menor disponibilidade de tempo. Como psicólogo, vejo que o burnout não aparece aqui porque quem não consegue equilibrar simplesmente abandona o trabalho (ou a faculdade).





### Padrão geral: Sim, há relação — filhos de pais com ensino superior (Bachelor) tendem a ir melhor. Porém, o resultado de pais com Mestrado ter nota média mais baixa (68,1) é um artefato amostral (grupo pequeno, N=167) ou pode refletir um perfil específico: pais acadêmicos muito exigentes geram mais pressão e ansiedade, ou esses alunos vêm de contextos migrantes com mais barreiras.

### Como psicólogo: A escolaridade dos pais opera via capital cultural (valorização dos estudos, ambiente intelectual em casa, expectativas) e capital econômico (acesso a recursos). O efeito existe, mas é mais sutil do que se imagina — a diferença entre Bachelor e High School é de menos de 1 ponto.

## 4. 😴 Existe uma quantidade ideal de sono para melhores notas?




### Correlação: r = 0,12 (fraca, mas p < 0,0001 — significativa pela amostra grande) ANOVA: F = 3,57, p = 0,0017 → diferenças entre grupos são significativas.

### 🎯 A "janela ideal": 7 a 8 horas de sono
### Quem dorme 7–8 horas tira em média 72,4 — a maior nota entre grupos com tamanho amostral robusto. Quem dorme menos de 4 horas vai muito mal (59,9, quase 13 pontos abaixo). Curiosamente, o grupo de 9–10 horas tem nota altíssima (75,7), mas é um grupo pequeno (N=22) e provavelmente inclui alunos que dormem muito por depressão, fadiga crônica ou insônia paradoxal — não por "super-recuperação".

### Como especialista em saúde do adolescente:

### O sono é o fator isolado mais impactante nesta análise (diferença de ~12 pontos entre extremos)
### 7–8h é a faixa fisiológica ideal para consolidar memória (fase REM) e regular neurotransmissores
### Menos de 6h compromete atenção, controle inibitório e regulação emocional
### Mais de 9h geralmente indica hipersônia reativa (compensando privação crônica ou associada a transtornos de humor)



### Recomendação prática para adolescentes: Priorize o sono (7–8h), mantenha uma dieta minimamente equilibrada (frutas, proteínas, hidratação), e saiba que o contexto familiar importa — mas não é destino. O trabalho de meio período, se bem gerenciado, não precisa prejudicar o desempenho.
