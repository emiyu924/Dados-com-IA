Programa que avalia as notas que o usuário do nosso aplicativo inseriu.


```python
avaliacao = float(input("Digite sua avaliação: (de 0 - 10 estrelas"))
if avaliacao > 8:
    print("Nota Excelente!")
elif avaliacao > 6:
    print("Avaliação regular")
elif avaliacao > 4:
    print("Avaliação ruim!")
else:
    print("Avaliação péssima!")
```

Sistema que irá verificar se o filme avaliado é bom ou não
- Criar uma variável que irá armazenar o nome do filme 
- Criar uma variável que irá armazenar a nota do filme (0 - 5)
- Condição: Se possuir 5 estrelas, filme Vencedor do Oscar
- Condição: se possuir de 3 a 4.99, filme Bom para assistir
- Abaixo disso, filme não indicado

```python
nome_filme = (input("Digite o nome do Filme a ser avaliado"))
avaliacao = float(input("Agora digite sua avaliação: (de 0 - 5 estrelas"))
if avaliacao > 5:
    print(f"Você estourou a escala de tão bom que é: {nome_filme}, parabéns para os criadores! ")
elif avaliacao == 5:
    print(f"{nome_filme} é praticamente um vencedor do Oscar, pode assistir que é sucesso!")
elif avaliacao >= 3:
    print(f"{nome_filme} é ótimo para assistir e matar o tempo!")
elif avaliacao < 0:
    print(f"{nome_filme} é tão ruim que você quer dar uma nota menor que zero, pode jogar fora")
else:
    print(f"Pode largar a tv e ir fazer outra coisa, {nome_filme} não é indicado!")
```
